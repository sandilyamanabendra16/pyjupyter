```python
df1=pd.DataFrame({'Rollno':[1,2,3,4],
                 'Name':['Prateek','Yogesh','Komal','Shalu'],
                 'Marks':[40,50,60.5,70],
                 'Gender':['M','M','F','F']})
print(df1)
type(df1)
```

```python
df1
```

```python
print('Column names',df1.columns)
print('Index/ Row names',df1.index)
print('Datatypes \n',df1.dtypes)
print('Shape of DF(R-C)',df1.shape)
```

```python
print('Describe the values')
df1.describe()
```

```python
 df1.groupby('Gender').size()  # how many male and females
```

```python
df1.groupby('Gender')['Marks'].mean()  # Mean marks of each gender type
```

```python
df1.groupby('Gender').aggregate({'Marks':[np.mean,'max','min','std','count']})
```
